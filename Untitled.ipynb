```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv('Service_Requests_Analysis.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['School or Citywide Complaint', 'Vehicle Type', 'Taxi Company Borough', 'Taxi Pick Up Location', 'Garage Lot Name'], axis = 1, inplace = True)
```

```python
df2 = df.groupby('Agency')['Complaint Type'].value_counts().to_frame().rename(columns = {'Complaint Type' : 'Count'}).sort_values('Count', ascending = False)
df2.reset_index(inplace = True)
df2
```

```python
fig = plt.figure(figsize = (10,8))
plt.grid()
plt.barh(df2['Complaint Type'], df2['Count'], label = 'complaint count', color = 'red', edgecolor = 'black', 
            linewidth = 1.8, alpha = 0.8)

for i, v in enumerate(df2['Count']):
    plt.text(v+0.2, i+.1, str(v), color = 'blue', fontweight = 'bold')
    
plt.show()
```

```python
# 10 kota dengan komplain terbanyak
city=df.groupby('City')['Complaint Type'].count().to_frame().sort_values('Complaint Type',ascending=False).reset_index()
city10 = city.iloc[:10]

fig = plt.figure(figsize = (8,8))
plt.pie(city10['Complaint Type'], labels = city10['City'], autopct = '%0.1f%%', radius = 1, shadow = True, 
           textprops = {'fontsize' : 15, 'color' : 'black'});

plt.show()
city10
```

```python
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])
df['Closing Time'] = df['Closed Date'] - df['Created Date']
df['Closing Time (minutes)'] = df['Closing Time'].dt.seconds/60
df.groupby(['Status'])['Closing Time (minutes)'].mean().to_frame()
```
